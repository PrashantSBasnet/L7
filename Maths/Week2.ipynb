Solving Linear Equation

x1 - x2 + x3 = 0 <br>
10x2- 25x3 = 90 <br>
20x1 + 10x2 = 80

```python
import numpy as np
```

```python
#matrix form
A = np.array([[1,-1,1],
              [0,10,-25],
              [20,10,0]])

b = np.array([0,90,80])
```

```python
x = np.linalg.solve(A,b)
```

```python
print (x)
```

```python
print (np.dot(A,x))
```

```python
# Let us confirm the solution.
# this shows one element is not equal because of float tolerance
print (np.dot(A,x) ==b)
```

```python
# here we use a tolerance comparison to show the differences is less
# than a defined tolerance.
TOLERANCE = 1e-12
print(np.abs((np.dot(A, x) - b)) <= TOLERANCE)
```
